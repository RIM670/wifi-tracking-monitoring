```python
import subprocess
import re
import time
import platform

# Function to read data from the command prompt
def read_data_from_cmd():
    if platform.system() == 'Linux':
        p = subprocess.Popen("iwconfig", stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    elif platform.system() == 'Windows':
        p = subprocess.Popen("netsh wlan show interfaces", stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    else:
        raise Exception('Unsupported platform')

    out = p.stdout.read().decode()
   
    # Regex for Linux
    if platform.system() == 'Linux':
        m = re.findall(r'(wlan[0-9]+) .*? Signal level=(-[0-9]+) dBm', out, re.DOTALL)
    # Regex for Windows
    elif platform.system() == 'Windows':
        m = re.findall(r'Signal\s*:\s*(\d+)%', out, re.DOTALL)
    else:
        raise Exception('Unsupported platform')

    p.communicate()

    return m

# Loop to get network strength
while(True):
    print('Network strength :', end='\n')
    data = read_data_from_cmd()
    print(data)
    time.sleep(4)
```
